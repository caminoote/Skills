```python
img = Image.open("C:/Users/skills/Desktop/training_datasets/matrix.png")
def change_contrast(img, factor):
    def contrast(pixel):
        return 128 + factor * (pixel - 128)
    return img.point(contrast)
img = np.array(img)
img = Image.fromarray(img.astype(np.uint8))
img1 = change_contrast(img, factor=1.5)
img1.save('C:/Users/skills/Desktop/training_datasets/result_cont.png')
```

```python
image = Image.open('C:/Users/skills/Desktop/training_datasets/matrix.png')
arr = np.asarray(image, dtype='uint8')
x, y, _ = arr.shape
 
k = np.array([[[0.2989, 0.587, 0.114]]])
arr2 = np.round(np.sum(arr*k, axis=2)).astype(np.uint8).reshape((x, y))
 
img2 = Image.fromarray(arr2)
img2.save('C:/Users/skills/Desktop/training_datasets/result_gray.png')
```
